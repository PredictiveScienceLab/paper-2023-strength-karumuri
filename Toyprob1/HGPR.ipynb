```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# get_ipython().run_line_magic('matplotlib', 'inline')
# import seaborn as sns
# sns.set() # Setting seaborn as default style even if use only matplotlib
import GPy
import argparse
import sys 
import os
os.environ['PYTHONHASHSEED'] = '0'
import seaborn as sns
sns.set_style("white")
plt.rcParams.update({'font.size': 12})
```

```python
seed = 20
```

```python
resultdir = os.path.join(os.getcwd(), 'HGPR_results' ,str(seed)+'-seed') 

if not os.path.exists(resultdir):
    os.makedirs(resultdir)
```

# Toy prob

```python
np.random.seed(seed=seed)
```

```python
g = lambda x: np.sin(8.0 * np.pi * x)
f = lambda x: (x - np.sqrt(2.0)) * g(x) ** 2
std_dev_g = 0.25
std_dev_f = 0.20
```

```python
''' Test data '''
X_p = np.linspace(0.0, 1.0, 150)[:,None]
z_p = g(X_p) + std_dev_g * np.random.RandomState(seed=seed).randn(*g(X_p).shape) # inexpensive measurements
y_p = f(X_p) + std_dev_f * np.random.RandomState(seed=seed).randn(*f(X_p).shape) # expensive measurements
```

```python
''' Train data '''
N_g = 65
N_f = 25

X_g = np.linspace(0, 1, N_g)[:,None]
z_g = g(X_g) + std_dev_g * np.random.RandomState(seed=seed).randn(*g(X_g).shape)

# for nested data
perm = np.random.RandomState(seed=seed).permutation(N_g)
X_f = X_g[perm[0:N_f]]
z_f = z_g[perm[0:N_f]]
y_f = f(X_f) + std_dev_f * np.random.RandomState(seed=seed).randn(*f(X_f).shape)

print (X_g.shape, z_g.shape)
print (X_f.shape, y_f.shape)


plt.figure(figsize=(14,5))

plt.subplot(1, 2, 1)
plt.plot(X_p, g(X_p), 'k--', label="True response g")
plt.scatter(X_g, z_g, s=20, color='darkorange', marker='o', label=r"Data $\mathbf{z}_g$")
plt.xlabel('$x$', fontsize=12)
# plt.ylabel('$y$', fontsize=12)
plt.legend(loc="best", fontsize=12)

plt.subplot(1, 2, 2)
plt.plot(X_p, f(X_p), 'k-', label='True response f')
plt.scatter(X_f, y_f, s=20, color='darkgreen', marker='x', label=r"Data $\mathbf{y}_f$")
plt.xlabel('$x$', fontsize=12)
# plt.ylabel('$y$', fontsize=12)
plt.legend(loc="best", fontsize=12)

plt.savefig(os.path.join(resultdir,str(seed)+'-'+'data.pdf'),dpi=300)
plt.show()
plt.close()
```

## HGPR

In this Case, we are going to build a MFGP by building two GPs

We build GP_g using {X=X_g, Y=z_g}, mean prediction = g_pred

We build GP_f using {X=[X_f,g_pred(X_f)], Y=y_f}, mean prediction = f_pred

where z_g, y_f are noisy data

```python
print ('Building GP_g between X_g and z_g')
kg = GPy.kern.RBF(X_g.shape[1], ARD=True)
mg = GPy.models.GPRegression(X_g, z_g, kg)
mg.Gaussian_noise.variance.constrain_fixed(std_dev_g**2)
mg.optimize_restarts()
print(mg)
print(mg[".*Gaussian_noise"])
print(mg[".*rbf"])

# Evaluate the previous fidelity model on the inputs of this fidelity
g_pred, _ = mg.predict(X_f) 

print ('Building GP_f between [X_f,g_pred] and y_f')
X_all = np.hstack((X_f,g_pred))
kf = GPy.kern.RBF(X_all.shape[1], ARD=True)
mf = GPy.models.GPRegression(X_all, y_f, kf)
mf.Gaussian_noise.variance.constrain_fixed(std_dev_f**2)
mf.optimize_restarts()
print(mf)
print(mf[".*Gaussian_noise"])
```

```python
def posterior_samples_f(X, size=100):
    # To store the samples
    As = []
    G = mg.posterior_samples_f(X, size=size) # The _f is required
    As.append(G)
    # To store samples from this fidelity
    Fsi = np.ndarray((X.shape[0], 1, size))
    for s in range(size):
        Xsi =  np.hstack([X, As[0][:, :, s]])
        # This is the computationally expensive step:
        Fsi[:, :, s] = mf.posterior_samples_f(Xsi, size=1)[:, :, 0]
    As.append(Fsi)
    return As
```

```python
print('MFGP Predictions: g_pred') # posterior (g*|z)
postsam_g = posterior_samples_f(X_p, size=1000)[0] # posterior_samples_f gives epistemic uncertainity only
print (postsam_g.shape)

mupred_g, varpred_g = np.mean(postsam_g, axis=2), np.var(postsam_g, axis=2)

print('MFGP Predictions: f_pred') # posterior (f*|y)
postsam_f = posterior_samples_f(X_p, size=1000)[1] # posterior_samples_f gives epistemic uncertainity only
print (postsam_f.shape)

mupred_f, varpred_f = np.mean(postsam_f, axis=2), np.var(postsam_f, axis=2)
```

```python
from sklearn.metrics import mean_absolute_error
def mean_relative_absolute_error(y_true, y_pred): 
    return np.mean( np.abs((y_pred-y_true)/y_true) )

print('GP Predictions plots')
z_MRelAE_wrt_truth = mean_relative_absolute_error( g(X_p), mupred_g )
z_MAE_wrt_truth = mean_absolute_error( g(X_p),  mupred_g )

z_MRelAE = mean_relative_absolute_error( z_p, mupred_g )
z_MAE = mean_absolute_error( z_p,  mupred_g )

print ('************************************************')
print ('z MRel.AE wrt truth: '+str(z_MRelAE_wrt_truth))
print ('z MAE wrt truth: '+str(z_MAE_wrt_truth ))

print ('************************')

print ('z MRel.AE: '+str(z_MRelAE))
print ('z MAE: '+str(z_MAE))
print ('************************************************')


fig, ax = plt.subplots(figsize=(6,5))
Xdata = [X_g]
Ydata = [z_g]
Yp = [mupred_g]
Vp = [varpred_g]
markers = ['o']
colors = ['darkorange']

for x, y, yp, vp, c, mk in zip(Xdata, Ydata, Yp, Vp, colors, markers):
    ax.plot(x, y, mk, color='black', markeredgewidth=1)
    ax.plot(X_p, yp, color=c)
    sp = np.sqrt(vp)
    lp = yp - 1.96 * sp
    up = yp + 1.96 * sp
    ax.fill_between(X_p.flatten(), lp.flatten(), up.flatten(), color=c,
                    alpha=0.15)
ax.plot(X_p, g(X_p), 'k-')
ax.set_xlabel('$x$', fontsize=12)
# ax.set_ylabel('$y$')
plt.tick_params(bottom=True, left=True)
plt.legend([r"Data $\mathbf{z}_g$", "Prediction g", "True response g"], loc='best')
plt.title('HGPR')

plt.savefig(os.path.join(resultdir,str(seed)+'-'+'z_pred.pdf'),dpi=300)
plt.show()
plt.close()

##############################################################################

from sklearn.metrics import mean_absolute_error
def mean_relative_absolute_error(y_true, y_pred): 
    return np.mean( np.abs((y_pred-y_true)/y_true) )

print('GP Predictions plots')
y_MRelAE_wrt_truth = mean_relative_absolute_error( f(X_p), mupred_f )
y_MAE_wrt_truth = mean_absolute_error( f(X_p),  mupred_f )

y_MRelAE = mean_relative_absolute_error( y_p, mupred_f )
y_MAE = mean_absolute_error( y_p,  mupred_f )

print ('************************************************')
print ('y MRel.AE wrt truth: '+str(y_MRelAE_wrt_truth))
print ('y MAE wrt truth: '+str(y_MAE_wrt_truth ))

print ('************************')

print ('y MRel.AE: '+str(y_MRelAE))
print ('y MAE: '+str(y_MAE))
print ('************************************************')


fig, ax = plt.subplots(figsize=(6,5))
Xdata = [X_f]
Ydata = [y_f]
Yp = [mupred_f]
Vp = [varpred_f]
markers = ['x']
colors = ['darkgreen']

for x, y, yp, vp, c, mk in zip(Xdata, Ydata, Yp, Vp, colors, markers):
    ax.plot(x, y, mk, color='black', markeredgewidth=1)
    ax.plot(X_p, yp, color=c)
    sp = np.sqrt(vp)
    lp = yp - 1.96 * sp
    up = yp + 1.96 * sp
    ax.fill_between(X_p.flatten(), lp.flatten(), up.flatten(), color=c,
                    alpha=0.25)
ax.plot(X_p, f(X_p), 'k-')
ax.set_xlabel('$x$', fontsize=12)
# ax.set_ylabel('$y$')
plt.tick_params(bottom=True, left=True)
plt.legend([r"Data $\mathbf{y}_f$", 'Prediction f', 'True response f'], loc='best')
plt.title('HGPR')

plt.savefig(os.path.join(resultdir,str(seed)+'-'+'y_pred.pdf'),dpi=300)
plt.show()
plt.close()
```

## Diagnostic checks

If the model is correct i.e., if the statistics predicted by model is correct, then ECDF evaluated at the test data should follow a uniform distribution.


k_i ~ F (z_p_i) ~ U[0,1]

where z_p is input test data 

k_i ~ F (y_p_i) ~ U[0,1]

where y_p is output test data 

```python
postsamples_g = postsam_g[:,0,:].T
postsamples_g = postsam_g[:,0,:].T
print(postsamples_g.shape, postsamples_g.shape)

postsamples_f = postsam_f[:,0,:].T
postsamples_f = postsam_f[:,0,:].T
print(postsamples_f.shape, postsamples_f.shape)
```

```python
sigma_g = np.sqrt(mg.Gaussian_noise.variance.values[0])
sigma_f = np.sqrt(mf.Gaussian_noise.variance.values[0])
print (sigma_g, sigma_f) 
```

```python
# predictive posterior (z*|z)
predpostsamples_g = np.zeros((postsamples_g.shape[0], postsamples_g.shape[1]))
import pymc3 as pm
for i in range(predpostsamples_g.shape[0]):
    predpostsamples_g[i,:] = pm.Normal.dist(mu=postsamples_g[i,:], 
                                                sigma=sigma_g).random()
print (predpostsamples_g.shape)
print (z_p.shape)

# https://www.statsmodels.org/stable/generated/statsmodels.distributions.empirical_distribution.ECDF.html
from statsmodels.distributions.empirical_distribution import ECDF

ecdf_z_p = []
for i in range(z_p.shape[0]):
    ecdf = ECDF(predpostsamples_g[:,i]) # Build ECDF with all samples
    ecdf_z_p.append(ecdf(z_p[i])[0]) # CDF evaluated at test points
    
    
fig, ax = plt.subplots(dpi=100)
ax.plot(ecdf_z_p, 'o', color='red', markersize=3)
ax.plot(np.arange(z_p.shape[0]), 0.0 * np.ones(z_p.shape[0]), 'r--')
ax.plot(np.arange(z_p.shape[0]), 1.0 * np.ones(z_p.shape[0]), 'r--')
ax.set_xlabel('$i$')
ax.set_ylabel('$ECDF \ at \ z \ test \ (k_i)$');
plt.savefig(os.path.join(resultdir,str(seed)+'-'+'ECDF_at_z_test.pdf'),dpi=300)
plt.show()
plt.close()


import statsmodels.api as sm
import scipy.stats as st
import pylab
sns.set_style("white")
fig, ax = plt.subplots(figsize=(5,4))
fig_ = sm.qqplot(np.asarray(ecdf_z_p), st.uniform, loc=0, scale=1, marker='.', markerfacecolor='r', markeredgecolor='r', markersize=3, alpha=1, ax=ax)
sm.qqline(fig_.axes[0], line='45', fmt='k--')
ax.set_title('Q-Q Plot (HGPR)', fontsize=12)
ax.set_xlabel('Uniform quantiles', fontsize=12)
ax.set_ylabel('k sample quantiles', fontsize=12)
ax.tick_params(bottom=True, left=True)
fig.savefig(os.path.join(resultdir,str(seed)+'-'+'q-q_plot_z_test.pdf'),dpi=300)
plt.show()
plt.close()


# https://www.geeksforgeeks.org/ml-kolmogorov-smirnov-test/
res = st.kstest(ecdf_z_p,"uniform")    
print(res) 

ks = ecdf_z_p
fig, ax = plt.subplots(dpi=100)
ps = np.linspace(-0.5, 1.5, 100)
ax.plot(ps, st.uniform.pdf(ps))
ax.hist(ks, density=True, alpha=0.5, bins = 30)
ax.set_xlabel('$k_i$')
plt.savefig(os.path.join(resultdir,str(seed)+'-'+'k_i_z_test.pdf'),dpi=300)
plt.show()
plt.close()
```
