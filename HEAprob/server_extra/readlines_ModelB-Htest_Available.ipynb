```python
import os

# Set path to the folder: (replace with your own filepath!):
folder = os.getcwd()

objective_list = []
for i in range(len(filename)):
    
    # Generate full filepath by combining the folder and filename
    fp = os.path.join(folder, filename[i])
    
    f1 = open(fp, "r")
    last_line = f1.readlines()[-1]
    print (last_line)
    
    s_array = last_line.split('Strength KS test:')
    objective_list.append(s_array[1])
    
    f1.close()

Strength_KS_test_list = list(map(float, objective_list))
print(Strength_KS_test_list)
```
